## Filling Missing Values

```python
df['customer_id'].fillna('Unknown', inplace=True)
df['region'].fillna('Unknown', inplace=True)
df['category'].fillna('Unknown', inplace=True)
df['payment_method'].fillna('Unknown', inplace=True)

df['customer_age'].fillna(df['customer_age'].median(), inplace=True)
df['rating'].fillna(df['rating'].mean(), inplace=True)
df['discount_pct'].fillna(0, inplace=True)
```

## Correcting Unit Price Datatype

```python
df['unit_price'] = (
    df['unit_price']
    .astype(str)
    .str.replace(',', '')  # remove commas first
    .str.replace('[^0-9.]', '', regex=True)
)

df['unit_price'] = pd.to_numeric(df['unit_price'], errors='coerce')
```

## Correcting Order_date & Order_time Datatypes

```python
df['order_date'].isnull().sum()
df['order_time'].isnull().sum()
```

```python
df['order_date'] = df['order_date'].astype(str).str.strip()
df['order_time'] = df['order_time'].astype(str).str.strip()
```

```python
df
```

```python
def clean_date(date):
    try:
        return pd.to_datetime(date, format='%Y/%m/%d')
    except:
        try:
            return pd.to_datetime(date, format='%d/%m/%Y')
        except:
            try:
                return pd.to_datetime(date, format='%m-%d-%Y')
            except:
                return pd.NaT

df['order_date_clean'] = df['order_date'].apply(clean_date)
```

```python
df
```

```python
df['order_datetime'] = pd.to_datetime(
    df['order_date_clean'].astype(str) + ' ' + df['order_time']
)
```

```python
df
```

```python
df['year'] = df['order_datetime'].dt.year
df['month'] = df['order_datetime'].dt.month
df['day'] = df['order_datetime'].dt.day
df['hour'] = df['order_datetime'].dt.hour
```

```python
df
```
